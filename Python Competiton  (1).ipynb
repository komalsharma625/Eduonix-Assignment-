```python
# Sample library inventory
library_inventory = {
    "Book 1": 5,
    "Book 2": 1,
    "Book 3": 7,
    "Book 4": 2,
    "Book 5": 4
}

def display_books():
    print("Available Books:")
    for book, quantity in library_inventory.items():
        print(f"{book} : {quantity} available")

def borrow_book():
    book_title = input("Enter the title of the book you want to borrow: ")
    if book_title in library_inventory:
        quantity_to_borrow = int(input(f"How many copies of '{book_title}' would you like to borrow? "))
        if quantity_to_borrow <= library_inventory[book_title] and quantity_to_borrow > 0:
            library_inventory[book_title] -= quantity_to_borrow
            if quantity_to_borrow == 1:
                print(f"You have borrowed '{book_title}' (1 copy). Enjoy reading!")
            else:
                print(f"You have borrowed '{book_title}' ({quantity_to_borrow} copies). Enjoy reading!")
        else:
            print("Sorry, the requested quantity is not available.")
    else:
        print("Sorry, the book is either unavailable or not in the inventory.")

def return_book():
    book_title = input("Enter the title of the book you want to return: ")
    if book_title in library_inventory:
        library_inventory[book_title] += 1
        print(f"Thank you for returning , You have returned the 1 copies of '{book_title}'.")
    else:
        print("Invalid book title or not from this library.")

def main():
    while True:
        print("\nLibrary Management System Menu:")
        print("1. Display available books")
        print("2. Borrow a book")
        print("3. Return a book")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            display_books()
        elif choice == '2':
            borrow_book()
        elif choice == '3':
            return_book()
        elif choice == '4':
            print("Exiting the Library Management System . Thank you!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
