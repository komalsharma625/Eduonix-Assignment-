# ATM Simulator Project

Objective :

The Objective of the project is to build an application which is exactly working as an ATM machine. You need to include the basic functionalities such as Check Balance, Withdraw, Deposit and Exit of an ATM Machine.

The input for the application from the user is to enter the PIN and then ask the user to select an appropriate option and perform it accordingly.

```python
balance = 1000
transaction = []

def check_bal() :
    print ('Available Balance is : {}'.format(balance))
    
def deposite(amount) :
    global balance
    if amount > 0 :
        balance += amount
        transaction.append('Deposited amount is {}'.format(amount))
        return '{} is deposited successfully '.format(amount)
    else :
        return 'Deposited amount is not valid !'
    
def withdraw(amount) :
    global balance 
    if 0 < amount <= balance :
        balance -= amount
        transaction.append('Withdraw amount is {}'.format(amount))
        return '{} is successfully withdraw '.format(amount)
    else :
        return 'Insufficient funds or invalid amount for withdrawal !'
    
def transaction_history() :
    return '\n'.join(transaction)

def check_pin(pin) :
    if pin == 2498 :
        return True
    else : 
        return False
    
def main() :
    pin = int(input('Enter your Pin : '))
    if not check_pin(pin) :
        print('Your Pin is incorrect')
        return
    
    
    while True :
        print("\nATM Simulator Menu:")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Transaction History")
        print("5. Exit")
        
        choice = int(input('Enter your choice: '))
        
        if choice == 1 :
            print(check_bal())
        elif choice == 2:
            amount = int(input('Enter deposited amount : '))
            print(deposite(amount))
        elif choice == 3:
            amount = int(input('Enter withdraw amount : '))
            print(withdraw(amount))
        elif choice == 4 :
            print('Transaction History : ')
            print(transaction_history())
        elif choice == 5 :
            print('Thanks for using ATM')
            break
        else :
            print('Invalid Input, Please try again !')
            
main()            
```

```python
# Second method by using class and create atleast 3 people with different pin and verify pin as according to person and 
 create people because every person has unique pin value 

class ATM:
    balance = 1000
    transaction = []

    def __init__(self, name, pin):
        self.name = name
        self.pin = pin

    def check_bal(self):
        print('Available Balance is: {}'.format(ATM.balance))

    def check_pin(self, entered_pin):
        return self.pin == entered_pin

    def deposite(self, amount):
        if amount > 0 :
            ATM.balance += amount
            ATM.transaction.append(f'{self.name} deposited ${amount}')
            return '{} is deposited successfully '.format(amount)
        else :
            return 'Deposited amount is not valid !'
            
            

    def withdraw(self, amount):
        if 0 < amount <= ATM.balance :
            ATM.balance -= amount
            ATM.transaction.append(f'{self.name} withdrawn ${amount}')
            return '{} is successfully withdraw '.format(amount)
        else :
            return 'Insufficient funds or invalid amount for withdrawal !'
        

    def transaction_history(self):
        for trans in ATM.transaction:
            print(trans)

    def main(self):
        entered_pin = int(input('Enter your PIN : '))
        if not self.check_pin(entered_pin):
            print('Your PIN is incorrect')
            return

        while True:
            print("ATM Simulator Menu:")
            print("1. Check Balance")
            print("2. Deposit")
            print("3. Withdraw")
            print("4. Transaction History")
            print("5. Exit")

            choice = int(input('Enter your choice: '))

            if choice == 1:
                print(self.check_bal())
            elif choice == 2:
                amount = int(input('Enter deposited amount: '))
                print(self.deposite(amount))
            elif choice == 3:
                amount = int(input('Enter withdraw amount: '))
                print(self.withdraw(amount))
            elif choice == 4:
                print('Transaction History:')
                print(self.transaction_history())
            elif choice == 5:
                print('Thanks for using ATM')
                break
            else:
                print('Invalid Input, Please try again!')

# Creating instances of ATM for different people with their respective PINs
person1 = ATM('Amit', 2349)
person2 = ATM('Shilpa', 4259)
person3 = ATM('Priya', 5678)

# as according to person 
person2.main()
```
